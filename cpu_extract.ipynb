```python
import os
import glob
import sys
import cv2
import argparse
from multiprocessing import Pool, current_process
```

```python
# Parameters
src_path = "..\Related Codes\Dataset\YouTube\YouTubeClips"
out_path = "dataset_msvd"
num_worker = 8
flow_type = "frame"
ext = "avi"
new_width = 224
new_height = 224
```

```python
new_size = (new_width, new_height)
```

```python
def dump_frames(vid_path):
    v_ = vid_path[0]
    video = cv2.VideoCapture(v_)
    vid_name = v_.split("\\")[-1].split(".")[-2]
    out_full_path = os.path.join(out_path, vid_name)

    fcount = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    
    try:
        os.mkdir(out_full_path)
    except OSError:
        pass
    
    file_list = []
    
    for i in range(fcount):
        ret, frame = video.read()
        resize_frame = cv2.resize(frame, (new_size[0], new_size[1]))
        assert ret
        cv2.imwrite("{}/frame_{:06d}.jpg".format(out_full_path, i), resize_frame)
        access_path = "{}/frame_{:06d}.jpg".format(vid_name, i)
        file_list.append(access_path)
        
    print("{} done".format(vid_name))
    sys.stdout.flush()
    
    return file_list
    
```

```python
if not os.path.isdir(out_path):
    print("Creating output folder: " + out_path)
    os.makedirs(out_path)
```

```python
vid_list = glob.glob(src_path + "/*." + ext)
```

```python
for i in range(len(vid_list)):
    print(vid_list[i])
    
print("Number of Videos: ", len(vid_list))
print("\(sss\)")
```

```python
pool = Pool(num_worker)
pool.map(dump_frames, zip(vid_list, range(len(vid_list))))
```
